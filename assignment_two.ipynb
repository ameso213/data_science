```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=adult_df, x='age', y='capitalGain')
plt.title('Scatter Plot of Age vs Capital Gains')
plt.xlabel('Age')
plt.ylabel('Capital Gains')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.countplot(data=adult_df, x='race')
plt.title('Count of Individuals by Race')
plt.xlabel('Race')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(25, 10)) # visualization
sns.countplot(data=adult_df, x='age')
plt.title('Count of Individuals by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
